<h2>Web Scraping The Grab Food Website</h2>

<h5>Import Required Libraries</h5>

```python
import requests
from bs4 import BeautifulSoup as bs
import numpy as np
import pandas as pd
```

```python
#Requesting the website
r = requests.get("https://food.grab.com/my/en/restaurants" )
soup = bs(r.content ,  from_encoding='utf-8')
print(soup.prettify())
```

<h5>Webscraping The Restaurant Name and Location</h5>

```python
rest_name = soup.select('h6.name___2epcT')
rest_name_loc = []
for food in rest_name:
    new_list= food.get_text()
    new_list = new_list.replace('\n' , " ")
    rest_name_loc.append(new_list)
print(rest_name_loc)
```

<h5>Webscraping The Restaurant Information</h5>

```python
restaurant_info = soup.find_all('div',attrs={"class" : "basicInfoRow___UZM8d cuisine___T2tCh"})
rest_info = []
for i in restaurant_info:
    rest_info.append(i.text)
print(rest_info)
```

<h5>Webscraping The Restaurant Ratings</h5>

```python
divs = soup.find_all('div',attrs={"class" : "numbersChild___2qKMV"})
rest_rats_dur_dist = []
for i in divs:
    rest_rats_dur_dist.append(i.text.replace('\xa0' , "").replace('•' , ""))
print(rest_rats_dur_dist)
```

```python
rating = []
for index ,  i in enumerate(rest_rats_dur_dist):
    if index%2 == 0 :
        rating.append(i)
print(rating)
```

<h5>Webscraping The Restaurant Preparation Duration and Distance</h5>

```python
distance = []
duration = []
for i in rest_rats_dur_dist:
    if i.endswith("km"):
        a , b = i.split("min")
        duration.append(a)
        distance.append(b)
```

```python
duration = [x + "mins" for x in duration]
```

```python
print(duration)
```

```python
print(distance)
```

<h5>Storing The Values In A Dataframe</h5>

```python
import pandas as pd
import numpy as np
data_frame = {}
data_frame['Restaurant Name and Location'] = rest_name_loc
data_frame['Restaurant Info'] = rest_info
data_frame['Rating'] = rating
data_frame['Preparation Duration'] = duration
data_frame['Distance From Home'] = distance
data_frame['Rating'] = rating
df = pd.DataFrame.from_dict(data_frame , orient = 'index')
df = df.transpose()
df.to_csv("df3.csv" , index = False)
```
